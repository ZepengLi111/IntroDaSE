```python
# 对源文件开头与结尾文字注释删除
import matplotlib.pyplot as plt
import numpy as np
data1 = np.loadtxt(open("daily_KP_SUN_2020.csv","rb"), delimiter=",",skiprows=0,usecols=[1,2,3])
total_hours_per_month = np.zeros(9)
average_hours_per_day = np.zeros(9)
count = 0
month = 1
for i in data1:
    if i[0] != month:
        average_hours_per_day[month-1] = total_hours_per_month[month-1]/count
        month += 1
        count = 0
    total_hours_per_month[int(i[0]-1)] += i[2]
    count += 1
average_hours_per_day[8] = total_hours_per_month[8]/30

plt.figure()
plt.subplot(211)
plt.bar(range(1,10), total_hours_per_month)

plt.subplot(212)
plt.bar(range(1,10), average_hours_per_day)
plt.show()
```

```python
import csv
Sepal_Length = [[],[],[]]
Sepal_Width = [[],[],[]]
Petal_Width = [[],[],[]]
Petal_Length = [[],[],[]]
with open('iris.csv') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        # 不同种类的特征分开收集，是为了以后画图时可以让不同种类的花的散点颜色不同
        if row['Species'] == 'setosa':
            Sepal_Length[0].append(float(row['Sepal_Length']))
            Sepal_Width[0].append(float(row['Sepal_Width']))
            Petal_Length[0].append(float(row['Petal_Length']))
            Petal_Width[0].append(float(row['Petal_Width']))
        if row['Species'] == 'versicolor':
            Sepal_Length[1].append(float(row['Sepal_Length']))
            Sepal_Width[1].append(float(row['Sepal_Width']))
            Petal_Length[1].append(float(row['Petal_Length']))
            Petal_Width[1].append(float(row['Petal_Width']))
        if row['Species'] == 'virginica':
            Sepal_Length[2].append(float(row['Sepal_Length']))
            Sepal_Width[2].append(float(row['Sepal_Width']))
            Petal_Length[2].append(float(row['Petal_Length']))
            Petal_Width[2].append(float(row['Petal_Width']))

plt.figure()
plt.subplot(231)
# 红色代表setosa，蓝色代表versicolor，黄色代表virginica。
plt.scatter(Sepal_Width[0], Sepal_Length[0], color='red', s=8, alpha=0.5)
plt.scatter(Sepal_Width[1], Sepal_Length[1], color='blue', s=8, alpha=0.5)
plt.scatter(Sepal_Width[2], Sepal_Length[2], color='yellow', s=8, alpha=0.5)

plt.subplot(232)
plt.scatter(Sepal_Width[0], Petal_Width[0], color='red', s=8, alpha=0.5)
plt.scatter(Sepal_Width[1], Petal_Width[1], color='blue', s=8, alpha=0.5)
plt.scatter(Sepal_Width[2], Petal_Width[2], color='yellow', s=8, alpha=0.5)

plt.subplot(233)
plt.scatter(Sepal_Width[0], Petal_Length[0], color='red', s=8, alpha=0.5)
plt.scatter(Sepal_Width[1], Petal_Length[1], color='blue', s=8, alpha=0.5)
plt.scatter(Sepal_Width[2], Petal_Length[2], color='yellow', s=8, alpha=0.5)

plt.subplot(234)
plt.scatter(Sepal_Length[0], Petal_Length[0], color='red', s=8, alpha=0.5)
plt.scatter(Sepal_Length[1], Petal_Length[1], color='blue', s=8, alpha=0.5)
plt.scatter(Sepal_Length[2], Petal_Length[2], color='yellow', s=8, alpha=0.5)

plt.subplot(235)
plt.scatter(Sepal_Length[0], Petal_Width[0], color='red', s=8, alpha=0.5)
plt.scatter(Sepal_Length[1], Petal_Width[1], color='blue', s=8, alpha=0.5)
plt.scatter(Sepal_Length[2], Petal_Width[2], color='yellow', s=8, alpha=0.5)

plt.subplot(236)
plt.scatter(Petal_Width[0], Petal_Length[0], color='red', s=8, alpha=0.5)
plt.scatter(Petal_Width[1], Petal_Length[1], color='blue', s=8, alpha=0.5)
plt.scatter(Petal_Width[2], Petal_Length[2], color='yellow', s=8, alpha=0.5)

plt.show()
############# 结论： 可以当使用Sepal_Width, Sepal_Length这两种特征画图时，versicolor与virginica（对应的蓝色和黄色）无法区分开。而对于任何其他的两种特征（其他五幅图）来说，setosa与其他种类的花是可以很容易的区分开的，而versicolor与virginica也可以区分开，只是在分界线周围会有或多或少的可以忽略不计的重叠区域。用Sepal_Width和Petal_Width以及Sepal_Length和Petal_Width区分效果最好。
```
